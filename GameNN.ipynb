```python
#Normalising the data between 0 and 1 dataframe['column_name'].min()
def normalise(a):
    for i in a:
        for j in a[i]:
            j = (j - a[i].min())/ (a[i].max() - a[i].min())
        return a
    
normalise(df)
```
